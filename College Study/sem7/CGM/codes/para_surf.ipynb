```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math

N = 10
u = np.linspace(0,1,N)
v = np.linspace(0,1,N)
U, V = np.meshgrid(u,v)

x = U
y = V
z = U**2+V**2

# If you don't have matplotlib comment below code
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x, y, z)
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
ax.set_zlabel("z axis")
plt.show()
```

```python
X = np.reshape(x,(1,np.size(x)))
Y = np.reshape(y,(1,np.size(y)))
Z = np.reshape(z,(1,np.size(z)))
R = np.array([[np.cos(3.14/4),np.sin(3.14/4),0],[-np.sin(3.14/4),np.cos(3.14/4),0],[0,0,1]])
I = np.vstack((X,Y,Z))
O = R.dot(I)
xx, yy, zz = np.vsplit(O,3)
XX = np.reshape(xx,np.shape(x))
YY = np.reshape(yy,np.shape(y))
ZZ = np.reshape(zz,np.shape(z))
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(XX, YY, ZZ)
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
ax.set_zlabel("z axis")
plt.show()
```
