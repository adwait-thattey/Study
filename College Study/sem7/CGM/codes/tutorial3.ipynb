```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math

N = 30
t = np.linspace(0,1,N)

x = np.cos(2*np.pi*t)
y = np.sin(2*np.pi*t)

dt = np.ones(N)#np.divide(np.ones(N),t^2+dx^2+dy^2)
dx = -2*np.pi*np.sin(2*np.pi*t)
dy = 2*np.pi*np.cos(2*np.pi*t)

normalizer = np.sqrt(dx*dx+dy*dy+dt*dt)

dt = dt/normalizer
dx = dx/normalizer
dy = dy/normalizer

# If you don't have matplotlib comment below code
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(t, x, y, 'gray')
ax.quiver3D(t, x, y, dt, dx, dy)
ax.scatter3D(t, x, y, c=y, cmap='hsv')
ax.set_xlabel("t axis")
ax.set_ylabel("x axis")
ax.set_zlabel("y axis")
plt.show()
```

```python
T = np.array([t**3, t**2, t, dt/dt])
TT = T.transpose()

P = np.array([[0,0],[1,0],[1,1],[0,1]])
```

```python
A = np.array([[0,0,0,1],[(1/3)**3,(1/3)**2,(1/3),1],[(2/3)**3,(2/3)**2,(2/3),1],[1,1,1,1]])
B = np.linalg.inv(A)

G = TT.dot(B)
fig = plt.figure()
plt.plot(t,G)
plt.show()
```

```python
Pt = G.dot(P)
fig=plt.figure()
plt.plot(Pt[:,0],Pt[:,1])
plt.show()
```
